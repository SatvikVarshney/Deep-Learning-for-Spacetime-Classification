```python
import pandas as pd

import numpy as np
```

```python
df = pd.read_csv("SingleFile.csv", sep=',', header=None)
```

```python
dfshuff = df.sample(frac=1,random_state=1)
dfshuff = dfshuff.reset_index(drop=True)
```

```python
dfshuff.to_csv('datashuffled.csv')
```

```python
#create some data with Names column
data = df
#create unique list of names
UniqueNames = data[5].unique()

#create a data frame dictionary to store your data frames
DataFrameDict = {elem : pd.DataFrame() for elem in UniqueNames}

for key in DataFrameDict.keys():
    DataFrameDict[key] = data[:][data[5] == key]
```

```python
dfII = DataFrameDict['II'].sample(frac=1,random_state=1)
dfI = DataFrameDict['I'].sample(frac=1,random_state=1)
dfIII = DataFrameDict['III'].sample(frac=1,random_state=1)
dfN = DataFrameDict['N'].sample(frac=1,random_state=1)
dfD = DataFrameDict['D'].sample(frac=1,random_state=1)
dfO = DataFrameDict['O'].sample(frac=1,random_state=1)
```

```python
length = 2000
equaldata = pd.concat([dfI[:126000], dfII[:90500], dfIII[:55500], dfN[:23500], dfD[:24500], dfO[:10000]]).reset_index()
equaldata = equaldata.drop("index",axis=1)
```

```python
equaldata.to_csv('equaldata.csv')
```

```python
dfshuff = equaldata.sample(frac=1,random_state=1)
dfshuff = dfshuff.reset_index(drop=True)
dfshuff.to_csv('shuffledratio.csv')
```
